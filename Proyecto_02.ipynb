**Regresion Logistica**

```python
# Datos de entrenamiento 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

url_airline = ['https://raw.githubusercontent.com/Benjamilior/Proyecto_version_2/master/ALUMNOS-trainData_0.csv',
               'https://raw.githubusercontent.com/Benjamilior/Proyecto_version_2/master/ALUMNOS-trainData_1.csv',
               'https://raw.githubusercontent.com/Benjamilior/Proyecto_version_2/master/ALUMNOS-trainData_2.csv',
               'https://raw.githubusercontent.com/Benjamilior/Proyecto_version_2/master/ALUMNOS-trainData_3.csv',
               'https://raw.githubusercontent.com/Benjamilior/Proyecto_version_2/master/ALUMNOS-trainData_4.csv',
               'https://raw.githubusercontent.com/Benjamilior/Proyecto_version_2/master/ALUMNOS-trainData_5.csv',
               'https://raw.githubusercontent.com/Benjamilior/Proyecto_version_2/master/ALUMNOS-trainData_6.csv',
               'https://raw.githubusercontent.com/Benjamilior/Proyecto_version_2/master/ALUMNOS-trainData_7.csv',
               'https://raw.githubusercontent.com/Benjamilior/Proyecto_version_2/master/ALUMNOS-trainData_8.csv',
               'https://raw.githubusercontent.com/Benjamilior/Proyecto_version_2/master/ALUMNOS-trainData_9.csv',]

#Merge CSV

airline = pd.concat(pd.read_csv(url_airline).assign(sourcefilename = url_airline)for url_airline in url_airline)
airline = airline.reset_index()
airline = pd.concat([airline])
airline = airline.reset_index()
airline
```

```python
#Datos de Evaluacion
airline_evaluation =pd.read_csv('https://raw.githubusercontent.com/Benjamilior/Proyecto_version_2/master/ALUMNOS-evalData.csv', header=0)
airline_evaluation
```

```python
#Transformando No-Show
airline['noshow4'] = [ 1 if s > 4  else 0 for s in airline['noshow']] 
airline.__getitem__("noshow4")
airline
```

```python
x = airline.p2p.array
x = x.reshape(-1,1)

y= airline.noshow4.array
y = y.reshape(-1,1)
```

```python
#Aplicando Regresion Logistica
from sklearn.linear_model import LogisticRegression
regresion_logistica = LogisticRegression()

#Entrenando la regresión logística con los datos de entrenamiento
regresion_logistica.fit(x,y)
```

```python
x_nuevo = airline_evaluation.p2p.array
x_nuevo = x_nuevo.reshape(-1,1)
x_nuevo
```

```python
#Realizando Prediccion con los nuevos datos
prediccion = regresion_logistica.predict(x_nuevo)
print(prediccion)
```

```python
betrade = pd.DataFrame(prediccion)
betrade
```
